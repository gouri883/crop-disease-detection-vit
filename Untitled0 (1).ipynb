```python
import os, shutil, random

RAW_DIR = "/content/plantvillage_subset"
OUT_DIR = "/content/data_split_subset"

os.makedirs(OUT_DIR, exist_ok=True)

def split_dataset(raw_dir=RAW_DIR, out_dir=OUT_DIR, split_ratio=0.8):
    classes = [d for d in os.listdir(raw_dir) if os.path.isdir(os.path.join(raw_dir, d))]
    for cls in classes:
        os.makedirs(f"{out_dir}/train/{cls}", exist_ok=True)
        os.makedirs(f"{out_dir}/val/{cls}", exist_ok=True)

        images = os.listdir(f"{raw_dir}/{cls}")
        random.shuffle(images)

        split = int(split_ratio * len(images))
        train_imgs = images[:split]
        val_imgs = images[split:]

        for img in train_imgs:
            shutil.copy(os.path.join(raw_dir, cls, img), f"{out_dir}/train/{cls}/{img}")
        for img in val_imgs:
            shutil.copy(os.path.join(raw_dir, cls, img), f"{out_dir}/val/{cls}/{img}")

    print("Dataset split completed!")

split_dataset()
```

```python
# ----- ViT fine-tune & evaluation (run in Colab) -----
# Install dependencies (only if not already installed)
!pip install -q timm==0.9.2 scikit-learn matplotlib seaborn

# ---------- imports ----------
import os, time, copy
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import transforms, datasets, models
import timm
from sklearn.metrics import confusion_matrix, classification_report

# ---------- Config (edit if you want) ----------
DATA_DIR = "/content/data_split_subset"   # <-- your split folder
MODEL_DIR = "/content/models_subset"
os.makedirs(MODEL_DIR, exist_ok=True)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
NUM_EPOCHS = 8           # adjust (5-10 is good for subset)
BATCH_SIZE = 16
IMG_SIZE = 224
LR = 3e-5
WEIGHT_DECAY = 1e-4
USE_AMP = True           # mixed precision if GPU available
PATIENCE = 4             # early stopping patience

# ---------- Data transforms & loaders ----------
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.7,1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(0.1,0.1,0.05),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

val_transforms = transforms.Compose([
    transforms.Resize(int(IMG_SIZE*1.14)),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

train_dir = os.path.join(DATA_DIR, "train")
val_dir   = os.path.join(DATA_DIR, "val")

train_dataset = datasets.ImageFolder(train_dir, transform=train_transforms)
val_dataset   = datasets.ImageFolder(val_dir, transform=val_transforms)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

num_classes = len(train_dataset.classes)
print("Device:", DEVICE)
print("Num classes:", num_classes)
print("Some classes:", train_dataset.classes[:10])

# ---------- helper functions ----------
def accuracy_top1(output, target):
    _, pred = output.topk(1, 1, True, True)
    pred = pred.squeeze(1)
    return pred.eq(target).sum().item()

def train_one_epoch(model, loader, criterion, optimizer, scaler=None):
    model.train()
    running_loss = 0.0
    running_corrects = 0
    total = 0
    for inputs, labels in tqdm(loader, desc="Train", leave=False):
        inputs = inputs.to(DEVICE)
        labels = labels.to(DEVICE)
        optimizer.zero_grad()
        if scaler is not None:
            with torch.cuda.amp.autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        running_loss += loss.item() * inputs.size(0)
        running_corrects += accuracy_top1(outputs, labels)
        total += inputs.size(0)
    return running_loss/total, running_corrects/total

def eval_model(model, loader, criterion):
    model.eval()
    running_loss = 0.0
    running_corrects = 0
    total = 0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in tqdm(loader, desc="Eval", leave=False):
            inputs = inputs.to(DEVICE)
            labels = labels.to(DEVICE)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * inputs.size(0)
            running_corrects += accuracy_top1(outputs, labels)
            total += inputs.size(0)
            preds = torch.argmax(outputs, dim=1).cpu().numpy()
            all_preds.append(preds)
            all_labels.append(labels.cpu().numpy())
    epoch_loss = running_loss/total
    epoch_acc = running_corrects/total
    if len(all_preds)>0:
        all_preds = np.concatenate(all_preds)
        all_labels = np.concatenate(all_labels)
    else:
        all_preds = np.array([])
        all_labels = np.array([])
    return epoch_loss, epoch_acc, all_preds, all_labels

# ---------- build model(s) ----------
def build_vit(num_classes, pretrained=True):
    model = timm.create_model('vit_base_patch16_224', pretrained=pretrained)
    if hasattr(model, 'head'):
        in_ch = model.head.in_features
        model.head = nn.Linear(in_ch, num_classes)
    else:
        model.reset_classifier(num_classes)
    return model

vit = build_vit(num_classes=num_classes, pretrained=True).to(DEVICE)

# ---------- training loop ----------
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(vit.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2, verbose=True)
scaler = torch.cuda.amp.GradScaler() if (USE_AMP and DEVICE.type=='cuda') else None

best_wts = copy.deepcopy(vit.state_dict())
best_acc = 0.0
epochs_no_improve = 0

history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[]}

for epoch in range(1, NUM_EPOCHS+1):
    print(f"\nEpoch {epoch}/{NUM_EPOCHS}")
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(vit, train_loader, criterion, optimizer, scaler=scaler)
    val_loss, val_acc, val_preds, val_labels = eval_model(vit, val_loader, criterion)
    t1 = time.time()
    print(f"Train loss: {train_loss:.4f} acc: {train_acc:.4f} | Val loss: {val_loss:.4f} acc: {val_acc:.4f} | time: {(t1-t0):.1f}s")
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(val_acc)

    scheduler.step(val_acc)

    if val_acc > best_acc:
        best_acc = val_acc
        best_wts = copy.deepcopy(vit.state_dict())
        torch.save(best_wts, os.path.join(MODEL_DIR, "vit_subset_best.pth"))
        epochs_no_improve = 0
    else:
        epochs_no_improve += 1

    if epochs_no_improve >= PATIENCE:
        print("Early stopping.")
        break

# load best
vit.load_state_dict(best_wts)
print("Best val acc:", best_acc)

# ---------- final eval ----------
val_loss, val_acc, preds, labels = eval_model(vit, val_loader, criterion)
print("\nFinal Eval - Loss: %.4f Acc: %.4f" % (val_loss, val_acc))
print("\nClassification report:")
print(classification_report(labels, preds, target_names=train_dataset.classes, zero_division=0))

# ---------- confusion matrix ----------
cm = confusion_matrix(labels, preds)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=False, cmap='Blues')
plt.title('Confusion matrix (ViT)')
plt.xlabel('Predicted'); plt.ylabel('True')
plt.show()

# ---------- save history (optional) ----------
import json
with open(os.path.join(MODEL_DIR, "history_subset.json"), "w") as f:
    json.dump(history, f)

print("Model, history saved in", MODEL_DIR)
```

```python
# ---------- ResNet50 baseline training ----------
import os, copy, time, numpy as np
from tqdm import tqdm
import torch, torch.nn as nn, torch.optim as optim
from torchvision import models
from sklearn.metrics import confusion_matrix, classification_report

# Config - change if needed
RESNET_MODEL_DIR = "/content/models_resnet"
os.makedirs(RESNET_MODEL_DIR, exist_ok=True)
NUM_EPOCHS_R = 6     # reduce to 4 for a quicker run
BATCH_SIZE_R = 16    # if OOM, set 8
LR_R = 1e-4

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device for ResNet:", DEVICE)

# build model
resnet = models.resnet50(pretrained=True)
in_ch = resnet.fc.in_features
resnet.fc = nn.Linear(in_ch, num_classes)  # num_classes from previous run
resnet = resnet.to(DEVICE)

# criterion, optimizer, scheduler
criterion_r = nn.CrossEntropyLoss()
optimizer_r = optim.AdamW(resnet.parameters(), lr=LR_R, weight_decay=1e-4)
scheduler_r = optim.lr_scheduler.ReduceLROnPlateau(optimizer_r, mode='max', factor=0.5, patience=2)

# reuse train_loader and val_loader you already created (they point to subset)
best_wts_r = copy.deepcopy(resnet.state_dict())
best_acc_r = 0.0
epochs_no_improve_r = 0
PATIENCE_R = 4

history_r = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[]}

def accuracy_top1_torch(output, target):
    _, pred = output.topk(1, 1, True, True)
    pred = pred.squeeze(1)
    return pred.eq(target).sum().item()

def train_one_epoch_r(model, loader, criterion, optimizer):
    model.train()
    running_loss = 0.0; running_corrects = 0; total = 0
    for inputs, labels_batch in tqdm(loader, desc="TrainResNet", leave=False):
        inputs = inputs.to(DEVICE); labels_batch = labels_batch.to(DEVICE)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels_batch)
        loss.backward(); optimizer.step()
        running_loss += loss.item() * inputs.size(0)
        running_corrects += accuracy_top1_torch(outputs, labels_batch)
        total += inputs.size(0)
    return running_loss/total, running_corrects/total

def eval_model_r(model, loader, criterion):
    model.eval()
    running_loss = 0.0; running_corrects = 0; total = 0
    all_preds = []; all_labels = []
    with torch.no_grad():
        for inputs, labels_batch in tqdm(loader, desc="EvalResNet", leave=False):
            inputs = inputs.to(DEVICE); labels_batch = labels_batch.to(DEVICE)
            outputs = model(inputs)
            loss = criterion(outputs, labels_batch)
            running_loss += loss.item() * inputs.size(0)
            running_corrects += accuracy_top1_torch(outputs, labels_batch)
            total += inputs.size(0)
            preds_np = torch.argmax(outputs, dim=1).cpu().numpy()
            all_preds.append(preds_np); all_labels.append(labels_batch.cpu().numpy())
    if len(all_preds)>0:
        all_preds = np.concatenate(all_preds); all_labels = np.concatenate(all_labels)
    else:
        all_preds = np.array([]); all_labels = np.array([])
    return running_loss/total, running_corrects/total, all_preds, all_labels

# Training loop
for epoch in range(1, NUM_EPOCHS_R+1):
    print(f"\n[ResNet] Epoch {epoch}/{NUM_EPOCHS_R}")
    t0 = time.time()
    tr_loss, tr_acc = train_one_epoch_r(resnet, train_loader, criterion_r, optimizer_r)
    val_loss_r, val_acc_r, val_preds_r, val_labels_r = eval_model_r(resnet, val_loader, criterion_r)
    t1 = time.time()
    print(f"Train loss: {tr_loss:.4f} acc: {tr_acc:.4f} | Val loss: {val_loss_r:.4f} acc: {val_acc_r:.4f} | time: {(t1-t0):.1f}s")
    history_r["train_loss"].append(tr_loss); history_r["train_acc"].append(tr_acc)
    history_r["val_loss"].append(val_loss_r); history_r["val_acc"].append(val_acc_r)
    scheduler_r.step(val_acc_r)

    if val_acc_r > best_acc_r:
        best_acc_r = val_acc_r
        best_wts_r = copy.deepcopy(resnet.state_dict())
        torch.save(best_wts_r, os.path.join(RESNET_MODEL_DIR, "resnet50_subset_best.pth"))
        epochs_no_improve_r = 0
    else:
        epochs_no_improve_r += 1
    if epochs_no_improve_r >= PATIENCE_R:
        print("Early stopping ResNet.")
        break

# load best weights and final eval
resnet.load_state_dict(best_wts_r)
val_loss_r, val_acc_r, preds_r, labels_r = eval_model_r(resnet, val_loader, criterion_r)
print("\nResNet Final - Loss: %.4f Acc: %.4f" % (val_loss_r, val_acc_r))
print("\nClassification report (ResNet):")
print(classification_report(labels_r, preds_r, target_names=train_dataset.classes, zero_division=0))

# Save ResNet results to OUT_DIR and Drive
OUT_DIR_R = "/content/results_resnet"
os.makedirs(OUT_DIR_R, exist_ok=True)
np.save(os.path.join(OUT_DIR_R, "resnet_preds.npy"), preds_r)
np.save(os.path.join(OUT_DIR_R, "resnet_labels.npy"), labels_r)
with open(os.path.join(OUT_DIR_R, "resnet_classification_report.txt"), "w") as f:
    f.write(classification_report(labels_r, preds_r, target_names=train_dataset.classes, zero_division=0))
cmr = confusion_matrix(labels_r, preds_r)
np.save(os.path.join(OUT_DIR_R, "resnet_confusion_matrix.npy"), cmr)
plt.figure(figsize=(12,10)); sns.heatmap(cmr, annot=True, fmt='d', cmap='Blues', xticklabels=train_dataset.classes, yticklabels=train_dataset.classes); plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR_R, "resnet_confusion_matrix.png")); plt.close()

# copy to Drive
try:
    drive_path = "/content/drive/MyDrive/PlantVillage_Experiment"
    !cp -r {OUT_DIR_R} {drive_path}
    !cp -r {RESNET_MODEL_DIR} {drive_path}
    print("Copied ResNet results to Drive:", drive_path)
except Exception as e:
    print("Could not copy ResNet results to Drive automatically:", e)
    print("ResNet results are in", OUT_DIR_R)
```
